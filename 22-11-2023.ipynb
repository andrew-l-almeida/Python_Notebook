```python
#generator in python

# A generator is a function that has the ability to pause and resume
# To define a generator we use parenthesis instead of square brackets

import sys
generator = (n for n in range(10000000))
iterable = [n for n in range(1000000)]
sys.getsizeof(iterable)
sys.getsizeof(generator)

#Generator functions
#We create a gererator function, using the yield
# Yield é como o return nas funcões, mas com uma diferença crucial. Ela pausa a execução da função e a cada vez que eu chamar o next() ele me retorna esse valor
# Yield is like the 'return' statment in functions, but with a crucial difference. It pauses the execution of the function and evety time I call next() it returns the value

def generator(n=0):
    print('This is a generator function because of the yield')
    yield 1
    print('Now, after I call next, it will return the second yield')
    yield 2
    print('One last time and I will finish...')
    yield 3
    return('Its over')

gen = generator()

next(gen)


# Yield from is used to continue other generator

def gen1():
    print('COMECOU GEN1')
    yield 1
    yield 2
    yield 3
    print('ACABOU GEN1')


def gen3():
    print('COMECOU GEN3')
    yield 10
    yield 20
    yield 30
    print('ACABOU GEN3')


def gen2(gen=None):
    print('COMECOU GEN2')
    if gen is not None:
        yield from gen
    yield 4
    yield 5
    yield 6
    print('ACABOU GEN2')


g1 = gen2(gen1())
g2 = gen2(gen3())
g3 = gen2()
for numero in g1:
    print(numero)
print()
for numero in g2:
    print(numero)
print()
for numero in g3:
    print(numero)
print()
```
